```python
lego_sets
```

```python
licensed_parent_theme=parent_theme[parent_theme["is_licensed"]==True]
licensed_parent_theme
```

```python
licensed_parent_theme["name"]=licensed_parent_theme["name"].apply(lambda x: x.strip())
a=pd.Series(licensed_parent_theme["name"])
dic={}
for i in a:
    dic[i]=True
dic
```

```python
lego_sets["is_licensed"]=lego_sets["parent_theme"].apply(lambda x:True if x in dic else False)
```

```python
lego_sets=lego_sets[~(lego_sets["set_num"].isna())]
lego_sets
```

```python
licensed_legosets=lego_sets[lego_sets["is_licensed"]==True]
```

```python
import seaborn as sns
from  matplotlib import pyplot as plt
```

```python
plt.pie(x=licensed_legosets["parent_theme"].value_counts(),labels=licensed_legosets["parent_theme"].unique(),\
        radius=3,rotatelabels=True,autopct="%.0f%%")
plt.show()
```
